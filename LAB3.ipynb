####**Program to compute Fibonacci numbers using the following 2 methods below:**

* *Memoization (Top Down) Fibonacci*

```python
def Memofibonnaci(n, cache=None):
    if cache is None:
        cache = {}
    if n <= 1:
        return n
    if not cache.get(n):
        cache[n] = Memofibonnaci(n - 1, cache) + Memofibonnaci(n - 2, cache)
    return cache[n]

print(f" Fibonacci(500) = {Memofibonnaci(500)}") # Faster than normal algorithm, runs n times only
```

 
* *Tabulation (Bottom Up) Fibonacci*

```python
def Tabfibonnaci(n):
    if n <= 1:
        return n
    cache = [0, 1]
    for i in range(2, n + 1):
        cache.append(cache[i - 1] + cache[i - 2])
    return cache[n]

print(f" Fibonacci(500) = {Tabfibonnaci(500)}") # Tabular Approach, Faster than normal algorithm, runs for n times only
```

#### The rod-cutting problem - Given a rod, length n inches and a table of prices. p[i] is the price of a rod of length i inches.

* ***Below is the Code to solve the rod-cutting problem using bottom-up approach***

```python
# Code
import numpy as np

def Bottom_Up_Tabular_Rod(p, n):
  r = {0:0}
  for j in range(1, n+1):
    q = -np.Inf
    for i in range(1, j+1):
      q = max(q, p[i]+ r[j-i])
    r[j] = q
  return r[n]

c = [1, 5, 8, 9, 10, 17, 17, 20, 24, 30]
print(Bottom_Up_Tabular_Rod(c, 4))
```
